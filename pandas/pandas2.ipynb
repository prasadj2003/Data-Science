#### Pandas Cleaning
1. Data cleaning means fixing bad data in your data set.

Bad data could be:
- Empty cells
- Data in wrong format
- Wrong data
- Duplicates

````python
import pandas as pd
````

````python
from datasets import load_dataset
````

````python
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
````

````python
df.info()
````

````python
df.job_posted_date[0]
type(df.job_posted_date[0])
````

````python
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
````

````python
df.info()
````

````python
df.job_posted_date.dt.day_of_week
````

````python
df.job_posted_month = df.job_posted_date.dt.month
````

````python
df['job_posted_month'] = df.job_posted_date.dt.month
````

````python
df.info()
````

#### pandas.DataFrame.sort_values: Sort the values along either axis
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html

````python
df.sort_values('job_posted_month', ascending=True, inplace=True)
df
````

#### Removing duplicates: 
```Python
print(df.duplicated())
```
Returns True for every row that is a duplicate, otherwise False

Solution: `df.drop_duplicates(inplace = True)` 

````python
df.drop_duplicates(inplace=True)
df.info()
````

#### To remove any columns with NaN values we will use the `DataFrame.drop` method

````python
df.drop(labels='salary_hour_avg', axis=1)
````

##### But if we run `df.info` the salary avg will still be present because we did not mention the inplace=True

#### The dropna() method removes the rows that contains NULL values. The dropna() method returns a new DataFrame object unless the inplace parameter is set to True, in that case the dropna() method does the removing in the original DataFrame instead.

````python
df.dropna(axis=1, inplace=True)
df.info()
````
